### Week1 项目报告：
工作概述：本周主要完成了项目的数据准备工作，包括数据需求分析（Data Requirement Specification）、API配置与测试（API Setup and Testing）以及初始数据收集（Initial Data Collection）。

数据需求分析：

1.金融市场数据：JPM（JPMorgan Chase）股票历史价格数据、VIX（Volatility Index）波动率指数数据

其中，JPM股票数据包含：Open（开盘价）、High（最高价）、Low（最低价）、Close（收盘价）、Adjusted Close（复权收盘价）、Volume（成交量）。VIX指数用于衡量市场预期波动率。

2.宏观经济数据：美国10年期国债收益率（DGS10）。

宏观经济数据主要用于反映市场利率环境。
